```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df["RD_Value"] = pd.to_numeric(df["RD_Value"], errors="coerce")
```

```python
print(df['Breakdown'].value_counts())
```

```python
df = df.dropna(subset=["RD_Value"])
```

```python
df["RD_Value"].describe()
```

```python
df["Year"].unique()
```

```python
avg_per_year = df.groupby("Year")["RD_Value"].mean()
plt.figure(figsize=(10,5))
plt.plot(avg_per_year.index, avg_per_year.values, marker="o")
plt.title("Average RD Value per Year")
plt.xlabel("Year")
plt.ylabel("Average RD_Value")
plt.grid(True)
plt.show()
```

```python
top_breakdown = df.groupby("Breakdown")["RD_Value"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
top_breakdown.plot(kind="bar", color="skyblue")
plt.title("Top 10 Breakdown categories by Avg RD_Value")
plt.xlabel("Breakdown")
plt.ylabel("Average RD_Value")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["RD_Value"], bins=30, color="orange", edgecolor="black")
plt.title("Distribution of RD_Value")
plt.xlabel("RD_Value")
plt.ylabel("Frequency")
plt.show()
```
